```python
#Importer les librairies necessaires pour notre traitement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

### **Explorer nos donn√©es**: üëÅ

```python
#Charger les donn√©es
df = pd.read_csv("/content/Import_Export.csv", delimiter=';')
display(df)
```

```python
df.info()
```

### *Effectuer une op√©ration de Pivot *

```python
# Utiliser la fonction `melt` pour convertir les colonnes 2005 √† 2022 en une seule colonne 'date'
df_melted = pd.melt(df, id_vars=['id_pays', 'Country', 'Subject Descriptor'], var_name='date', value_name='valeur')

# Utiliser la fonction `pivot` pour obtenir les types d'investissement comme colonnes
df = df_melted.pivot_table(index=['id_pays', 'Country', 'date'], columns='Subject Descriptor', values='valeur', aggfunc='sum').reset_index()
```

```python
df = df[['id_pays', 'Country', 'date', 'Volume of imports of goods and services', 'Volume of exports of goods and services']]
display(df)
```

# **G√©rer les donn√©es NaN**

```python
# Compter les NaN dans l'ensemble du DataFrame
total_nan = df.isna().sum().sum()

# Compter les NaN par colonne
nan_by_column = df.isna().sum()

# Afficher les r√©sultats
print(f"Nombre total de NaN dans le DataFrame : {total_nan}")
print("\nNombre de NaN par colonne :")
print(nan_by_column)
```
