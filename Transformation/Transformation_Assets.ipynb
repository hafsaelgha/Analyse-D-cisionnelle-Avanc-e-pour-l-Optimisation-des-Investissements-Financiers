```python
#Importer les librairies necessaires pour notre traitement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

### **Explorer nos donn√©es**: üëÅ

```python
#Charger les donn√©es
df = pd.read_csv("/content/Assets.csv", delimiter=',')
display(df)
```

```python
df.info()
```

**This data report uses the IIP Standard Presentation format as defined in the 6th Edition of the Balance of Payments Manual (BPM6).**



---


*   (-) Indicates that a figure is zero
*   (...) Indicates a lack of statistical data that can be reported or calculated from underlying observations
*   (K) indicates the splice point between official BPM6-basis estimates and IMF converted BPM5-basis estimates



```python
#Missing values & null values & (K)
years = [2004 + i for i in range(1,18)]
for y in years:
  df[str(y)] = df[str(y)].replace('...', None)
  df[str(y)] = df[str(y)].replace('-', 0)
  df[str(y)] = df[str(y)].str.replace('K', '').replace(',', '')
```

```python
display(df)
```

**Extraire les KPI qui nous int√©ressent et supprimer les lignes qui n'impacteront pas notre √©tude**

```python
# Identifier les lignes dupliqu√©es pour "Equity and investment fund shares"
duplicates_mask = (df['Investment Type'] == "Equity and investment fund shares") & df.duplicated(subset=['Investment Type'], keep='first')

# Supprimer les lignes correspondantes
df = df[~duplicates_mask]

# R√©initialiser les index si n√©cessaire
df.reset_index(drop=True, inplace=True)

# Supprimer les lignes avec les types d'investissement sp√©cifi√©s
types_a_supprimer = ["Debt instruments", "Central bank", "General government", "Debt securities"]
df = df[~df['Investment Type'].isin(types_a_supprimer)]

# R√©initialiser les index si n√©cessaire
df.reset_index(drop=True, inplace=True)
```

```python
display(df)
```

### *Effectuer une op√©ration de Pivot *

```python
# Utiliser la fonction `melt` pour convertir les colonnes 2005 √† 2022 en une seule colonne 'date'
df_melted = pd.melt(df, id_vars=['id_pays', 'Country', 'Investment Type'], var_name='date', value_name='valeur')

# Utiliser la fonction `pivot` pour obtenir les types d'investissement comme colonnes
df = df_melted.pivot_table(index=['id_pays', 'Country', 'date'], columns='Investment Type', values='valeur', aggfunc='sum').reset_index()
```

```python
display(df)
```

# **Convertir nos donn√©es en format num√©rique**

```python
# Supprimer les espaces dans les colonnes num√©riques
numeric_columns = ['Direct investment', 'Direct investor in direct investment enterprises', 'Equity and investment fund shares', 'Financial derivatives (other than reserves) and employee stock options', 'Other investment', 'Portfolio investment', 'Reserve assets']

for col in numeric_columns:
    df[col] = df[col].str.replace(' ', '')
    df[col] = df[col].str.replace(',', '.')
    df[col] = pd.to_numeric(df[col], errors='coerce')
#convertir en datetime
df['date'] = pd.to_datetime(df['date'], format='%Y')
df['date'] = df['date'].dt.year

df.info()
```

```python
display(df)
```

# **G√©rer les donn√©es NaN**

```python
# Compter les NaN dans l'ensemble du DataFrame
total_nan = df.isna().sum().sum()

# Compter les NaN par colonne
nan_by_column = df.isna().sum()

# Afficher les r√©sultats
print(f"Nombre total de NaN dans le DataFrame : {total_nan}")
print("\nNombre de NaN par colonne :")
print(nan_by_column)
```

```python
# Remplacer les NaN par la moyenne de la colonne
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col].fillna(df[col].mean(), inplace=True)
        df[col] = df[col].round(2)
display(df)
```

```python
#Enregistrer le dataframe transform√© en fichier csv
df.to_csv('trasformed_Assets.csv', index=False)
```
